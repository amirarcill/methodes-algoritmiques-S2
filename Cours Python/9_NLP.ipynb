# Introduction au Natural Language Processing (NLP)

## 1. Qu'est-ce que le NLP ?

Le **Natural Language Processing (NLP)**, ou **traitement automatique du langage naturel**, est un domaine de l’intelligence artificielle qui vise à permettre aux ordinateurs **de comprendre, analyser et produire du langage humain**.

Le langage naturel comprend toutes les formes de communication utilisées par les humains, par exemple :

- textes (articles, livres, posts sur les réseaux sociaux)
- conversations
- commentaires en ligne
- documents administratifs
- transcriptions d’entretiens

Le défi principal du NLP est que le langage humain est **ambigu, contextuel et extrêmement varié**.

Par exemple :

- *“This movie is good”* → sentiment positif  
- *“This movie is not good”* → sentiment négatif  

Pour un humain, la différence est évidente. Pour une machine, elle nécessite des méthodes spécifiques d’analyse linguistique et statistique.

# 4. Évolution historique du NLP et de l’intelligence artificielle

Le développement du **Natural Language Processing (NLP)** est étroitement lié à l’évolution plus générale de **l’intelligence artificielle (IA)**.  
On peut distinguer plusieurs grandes phases historiques, correspondant à différentes conceptions de l’intelligence et du langage.

---

## 1. L’IA symbolique (années 1950–1980)

Les premières approches de l’intelligence artificielle reposaient sur une idée centrale :  
**l’intelligence peut être décrite à l’aide de règles logiques explicites**.

Cette approche est appelée **IA symbolique**.

Dans ce cadre, le langage est traité comme un système de **règles grammaticales et syntaxiques** que l’on peut formaliser.

Les premiers systèmes de NLP étaient donc basés sur :

- des **dictionnaires**
- des **règles linguistiques**
- des **grammaires formelles**

Les chercheurs tentaient de modéliser explicitement la structure du langage.

### Exemple célèbre : ELIZA

Un exemple historique est le programme **ELIZA**, développé en 1966 par Joseph Weizenbaum.

ELIZA simulait un psychothérapeute en reformulant les phrases de l’utilisateur à l’aide de **règles simples de transformation**.

Exemple :

Utilisateur :

> I feel unhappy today.

ELIZA :

> Why do you feel unhappy today?

Le système ne comprenait pas réellement le langage : il appliquait simplement **des règles de substitution de mots et de structures**.

Limite principale de cette approche :

- difficulté à écrire toutes les règles nécessaires
- incapacité à gérer l’ambiguïté et la variété du langage naturel.

---

## 2. L’IA connexionniste et les approches statistiques (années 1990–2010)

À partir des années 1990, un changement important se produit :  
les chercheurs commencent à utiliser **des méthodes statistiques et des données massives** plutôt que des règles écrites à la main.

Cette approche est souvent associée à l’**IA connexionniste**, inspirée du fonctionnement des **réseaux de neurones**.

Dans le NLP, cela se traduit par l’utilisation de :

- modèles probabilistes
- apprentissage automatique (machine learning)
- grands corpus de texte

Le langage n’est plus décrit par des règles explicites, mais par **des régularités statistiques observées dans les données**.

Par exemple, on peut estimer la probabilité qu’un mot apparaisse après un autre mot.

C’est aussi durant cette période que se développent des représentations numériques du langage comme :

- **Bag-of-Words**
- **TF-IDF**
- **word embeddings** (Word2Vec, GloVe)

Ces méthodes permettent de représenter les mots et les textes sous forme de **vecteurs numériques**.

---

## 3. L’ère du deep learning et des transformers (années 2010–)

À partir des années 2010, les progrès du **deep learning** et l’augmentation de la puissance de calcul permettent de développer des modèles beaucoup plus complexes.

Une avancée majeure est l’introduction de l’architecture **Transformer** en 2017.

Ces modèles peuvent analyser les phrases **dans leur contexte complet**, plutôt que mot par mot.

Ils sont capables de :

- comprendre des relations complexes entre les mots
- capturer le contexte d’une phrase
- apprendre à partir de très grands corpus de texte.

---

## 4. Les grands modèles de langage (LLM)

Les modèles les plus récents sont appelés **Large Language Models (LLM)**.

Ils sont entraînés sur **des milliards ou des trillions de mots** et peuvent accomplir une grande variété de tâches linguistiques :

- répondre à des questions
- résumer des textes
- traduire
- générer du texte
- analyser des documents.

Contrairement aux systèmes plus anciens, ces modèles ne sont pas conçus pour une seule tâche spécifique.  
Ils apprennent une **représentation générale du langage** qui peut ensuite être utilisée dans de nombreux contextes.

---

# Représenter le langage sous forme numérique
Pour qu’un ordinateur puisse analyser un texte, il est nécessaire de **le transformer en données numériques**.  
Les ordinateurs ne peuvent pas interpréter directement les mots ou les phrases : ils manipulent uniquement **des nombres**.

Le traitement automatique du langage consiste donc, dans un premier temps, à **convertir le texte en représentations numériques** sur lesquelles des algorithmes peuvent travailler.

## Comment représenter un texte pour un ordinateur ?

Pour qu’un ordinateur puisse analyser un texte, il est nécessaire de **le transformer en données numériques**.  
Les ordinateurs ne peuvent pas interpréter directement les mots ou les phrases : ils manipulent uniquement **des nombres**.

Le processus se décompose en deux grandes étapes : 

---

### 1.Prétraitement du texte (Text preprocessing)

Avant de convertir le texte en nombres, il est souvent utile de **le nettoyer et de le préparer** :

- **Tokenisation** : découper le texte en unités élémentaires appelées *tokens* (souvent des mots, parfois des sous-mots ou des caractères).  
  Exemple : "J'aime le NLP" → ["J'", "aime", "le", "NLP"]

- **Nettoyage du texte** : supprimer les éléments inutiles comme la ponctuation, les caractères spéciaux ou les chiffres selon le contexte.

- **Normalisation** : mettre les mots sous une forme comparable, par exemple :  
  - passage en minuscules : "Paris" → "paris"  
  - lemmatisation ou stemming : "manges" → "manger"

- **Suppression des stopwords** : retirer les mots très fréquents qui n’apportent pas beaucoup d’information (ex. : *le, la, de, et*).

Ces étapes permettent de **préparer le texte** afin que sa représentation numérique soit plus efficace et pertinente.

---

### 2.Représentation numérique

Une fois le texte prétraité, il faut le **convertir en vecteurs ou matrices de nombres** que les algorithmes peuvent utiliser :

- **Bag-of-Words (BoW)** : chaque document est représenté par un vecteur indiquant le nombre d’occurrences de chaque mot du corpus.  
- **TF-IDF (Term Frequency - Inverse Document Frequency)** : pondère les mots selon leur fréquence dans le document et leur rareté dans le corpus.  
- **Embeddings / vecteurs distribués** : chaque mot ou phrase est représenté par un vecteur dense qui capture son sens et son contexte (Word2Vec, GloVe, FastText).  
- **Représentations contextuelles** : les modèles comme BERT ou GPT produisent des vecteurs qui tiennent compte du contexte de chaque mot dans la phrase.

Ces transformations permettent de passer **d’un texte lisible par un humain à une représentation mathématique exploitable par des modèles d’apprentissage automatique**.

## Nettoyage d’une liste de textes (Corpus preprocessing)

Dans la plupart des analyses NLP, on ne travaille pas sur une seule phrase mais sur **une liste de documents ou d’articles (corpus)**.

Le processus de nettoyage et de prétraitement se fait **document par document**. 

Pour automatiser la procédure, on regroupe toutes les étapes dans une seule fonction.

```python
# initialisation
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    # minuscules
    text = text.lower()
    # suppression de la ponctuation
    text = re.sub(r'[^\w\s]', '', text)
    # tokenisation
    tokens = word_tokenize(text)
    # suppression des stopwords
    tokens = [t for t in tokens if t not in stop_words]
    # lemmatisation
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    return tokens


````python
stop_words = set(stopwords.words('french'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    # minuscules
    text = text.lower()
    # suppression de la ponctuation
    text = re.sub(r'[^\w\s]', '', text)
    # tokenisation
    tokens = word_tokenize(text)
    # suppression des stopwords
    tokens = [t for t in tokens if t not in stop_words]
    # lemmatisation
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    return tokens
````

# Représentation numérique - Feature extraction

## Le modèle Bag-of-Words (BoW)

Le **modèle Bag-of-Words (BoW)** est l’une des méthodes les plus simples et les plus anciennes pour représenter un texte sous forme numérique en **traitement automatique du langage (NLP)**.

Le principe est de considérer un document comme un **“sac de mots”** (*bag of words*), c’est-à-dire une collection de mots **sans tenir compte de leur ordre dans la phrase**.  

Pour construire cette représentation, on procède généralement en deux étapes :

1. On construit un **vocabulaire** contenant tous les mots présents dans un corpus de textes.
2. Chaque document est représenté par un **vecteur** indiquant combien de fois chaque mot du vocabulaire apparaît dans le document.

Par exemple, si le vocabulaire est :

economy | jobs | inflation | market

et que le texte contient :

[economy,economy,jobs]

sa représentation Bag-of-Words sera :

[2,1,0,0]




### Exemple avec plusieurs textes

Supposons que nous avons trois courts textes :

Text 1: The economy is growing  
Text 2: The economy is slowing down  
Text 3: Jobs are growing in the tech sector  

Après un prétraitement simple (par exemple : minuscules, suppression de la ponctuation et des mots très fréquents), on peut construire un **vocabulaire** à partir de tous les mots présents dans le corpus :

economy | growing | slowing | jobs | tech | sector

Chaque texte est ensuite représenté par un vecteur indiquant **le nombre d’occurrences de chaque mot du vocabulaire** dans le document.

On obtient alors une **matrice document-terme** (*document-term matrix*) :

| Document | economy | growing | slowing | jobs | tech | sector |
|----------|--------|--------|--------|------|------|--------|
| Text 1   | 1 | 1 | 0 | 0 | 0 | 0 |
| Text 2   | 1 | 0 | 1 | 0 | 0 | 0 |
| Text 3   | 0 | 1 | 0 | 1 | 1 | 1 |

Dans cette matrice :

- chaque **ligne correspond à un document**
- chaque **colonne correspond à un mot du vocabulaire**
- chaque **valeur indique combien de fois le mot apparaît dans le document**


la bibliothèque **scikit-learn** propose une classe CountVectorizer pour créer la matrice "bag of words" (sac de mots)

````python
from sklearn.feature_extraction.text import CountVectorizer
````

````python
# generation de la liste de textes prétraités
texts=df['contenu_tweet'].tolist()
````

````python
french_stopwords = stopwords.words('french') # créer une liste de stopwords en français
vectorizer = CountVectorizer(stop_words=french_stopwords) # créer une instance de CountVectorizer en spécifiant les stopwords à supprimer

bow_matrix = vectorizer.fit_transform(texts) # convertir les listes de tokens en chaînes de caractères et créer la matrice de caractéristiques

print(bow_matrix.shape) # afficher la forme de la matrice de caractéristiques
````

La matrice Bag-of-Words a une dimension **400 × 48**.

Cela signifie que **chaque ligne correspond à un document** du corpus et **chaque colonne correspond à un mot différent du vocabulaire** extrait des textes.

- **400** : nombre de documents (par exemple 400 textes, commentaires ou phrases).
- **48** : nombre de mots distincts conservés dans le vocabulaire après le prétraitement du texte.

La valeur dans chaque cellule de la matrice indique **combien de fois un mot apparaît dans un document donné**.  
Par exemple, la valeur à la ligne *i* et à la colonne *j* représente **le nombre d’occurrences du mot *j* dans le document *i***.

````python
# afficher les features (mots unique du vocabulaire)
feature_names = vectorizer.get_feature_names_out()
print(feature_names)
````

````python
#afficher la matrice de caractéristiques
print(bow_matrix.toarray())
````

## TF-IDF (Term Frequency – Inverse Document Frequency)

La représentation **TF-IDF** est une évolution du modèle Bag-of-Words.  
Au lieu de simplement compter combien de fois un mot apparaît dans un document, TF-IDF attribue **un poids** à chaque mot en fonction de son importance dans le corpus.

L’idée est la suivante :

- **TF (Term Frequency)** mesure combien de fois un mot apparaît dans un document.
- **IDF (Inverse Document Frequency)** réduit l’importance des mots qui apparaissent dans **beaucoup de documents**, car ils sont généralement peu informatifs.

Ainsi, un mot aura un **poids élevé** s’il apparaît :
- souvent dans un document donné  
- mais **rarement dans les autres documents**.

À l’inverse, les mots très fréquents dans tout le corpus (par exemple *le*, *et*, *dans*, etc.) auront un **poids faible**.

La matrice TF-IDF a la **même structure qu’une matrice Bag-of-Words** :
- chaque **ligne** représente un document
- chaque **colonne** représente un mot du vocabulaire

La différence est que les cellules ne contiennent plus des **comptages**, mais des **poids TF-IDF**, qui reflètent l’importance relative des mots dans les documents.

La formule pour calculer le **TF-IDF** pour un mot \(t\) dans un document \(d\) est la suivante :

$$
\text{TF-IDF}(t,d) = \text{TF}(t,d) \times \text{IDF}(t)
$$

où :

- $\text{TF}(t,d) = \frac{\text{nombre d'occurrences de } t \text{ dans } d}{\text{nombre total de mots dans } d}$

- $\text{IDF}(t) = \log \frac{N}{1 + n_t}$

avec :
- $N$ = nombre total de documents dans le corpus  
- $n_t$ = nombre de documents contenant le mot $t$  

💡 La valeur **IDF** est grande pour les mots rares et faible pour les mots fréquents dans le corpus, ce qui permet de pondérer l’importance des mots.

````python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words=french_stopwords) # créer une instance de TfidfVectorizer en spécifiant les stopwords à supprimer
tfidf_matrix = vectorizer.fit_transform(texts) # convertir les listes de tokens en chaînes de caractères et créer la matrice de caractéristiques TF-IDF
print(tfidf_matrix.shape) # afficher la forme de la matrice de caractéristiques TF-IDF
````

````python
feature_names = vectorizer.get_feature_names_out()
print(feature_names)
# les caracteristiques (features) sont les meme mots que pour la matrice de caractéristiques "bag of words", 
# mais les valeurs dans la matrice TF-IDF sont pondérées en fonction de la fréquence des mots dans les documents et dans l'ensemble du corpus.
````

````python
#afficher la matrice de caractéristiques
print(tfidf_matrix.toarray())
````

### Word2Vec

**Word2Vec** est une technique d’**embeddings de mots** qui représente chaque mot par un **vecteur dense** dans un espace continu.  
Contrairement à Bag-of-Words ou TF-IDF, qui se basent sur des comptages ou des poids, Word2Vec capture le **sens et les relations sémantiques** entre les mots.

- Chaque mot est transformé en un vecteur de dimension fixe (par exemple 100 ou 300).  
- Les mots qui apparaissent dans des contextes similaires auront des vecteurs **proches** dans l’espace vectoriel.  

Il existe deux architectures principales pour entraîner Word2Vec :  
1. **CBOW (Continuous Bag-of-Words)** : prédit un mot à partir de son contexte.  
2. **Skip-gram** : prédit le contexte à partir d’un mot donné.  

💡 Exemple d’utilisation :  
- `roi - homme + femme ≈ reine`  
ce qui montre que Word2Vec capture des relations sémantiques et analogies entre mots.

#### Word2Vec et modèles pré-entraînés

Dans la pratique, on utilise souvent des **modèles Word2Vec pré-entraînés** plutôt que de former le modèle sur son propre petit corpus.  

- Ces modèles ont été entraînés sur des **corpus très larges** (par exemple Wikipédia, Google News, Common Crawl).  
- Ils contiennent des **représentations vectorielles riches** pour des millions de mots, capturant de nombreuses relations sémantiques et contextuelles.  
- On peut ensuite **utiliser ces vecteurs tels quels** pour représenter les mots de son propre texte, même si le corpus est petit, ou les **affiner (fine-tuning)** sur un corpus spécifique pour améliorer la pertinence.  

💡 Avantages :  
- Gain de temps et de ressources, pas besoin d’entraîner un modèle à partir de zéro.  
- Les vecteurs sont généralement **de meilleure qualité** et capturent des nuances que de petits corpus ne pourraient pas fournir.

Pour utiliser **Word2Vec** en pratique, on s’appuie souvent sur la bibliothèque **Gensim**, qui fournit des outils pour :  

- charger des **modèles pré-entraînés** (sur Wikipédia, Google News, etc.)  
- entraîner de nouveaux modèles sur un corpus spécifique  
- manipuler les vecteurs de mots pour calculer similarités ou représenter des textes  

💡 **Remarque importante :** contrairement à `CountVectorizer` ou `TfidfVectorizer`, Gensim **ne fait pas le pré-traitement automatiquement**.  
Il faut donc **passer au modèle un texte déjà pré-processé** :  


````python
# un exemple de word embeddings pré-entraîné en anglais
import gensim.downloader as api

# Charger un modèle Word2Vec pré-entraîné en anglais
model = api.load("word2vec-google-news-300") #importer le modele prends du temps, car le modèle est assez volumineux (environ 1,5 Go)
# mais une fois téléchargé, le modèle est stocké localement et peut être chargé rapidement à l'avenir.

# exemple d'utilisation du modèle de word embeddings
word1 = 'sociology'
vector1 = model[word1]
print(f"Le vecteur de '{word1}' est : {vector1}")
````

````python
word2='economics'
vector2 = model[word2]

word3='physics'
vector3 = model[word3]

from scipy.spatial.distance import cosine

def cosine_distance(vector_word1, vector_word2):
    return cosine(vector_word1, vector_word2)

print("Distanze (cosine distance) tra le parole:")
print(f"sociology - economics : {cosine_distance(vector1, vector2):.4f}")
print(f"sociology - physics : {cosine_distance(vector1, vector3):.4f}")
print(f"economics - physics : {cosine_distance(vector2, vector3):.4f}")
````
