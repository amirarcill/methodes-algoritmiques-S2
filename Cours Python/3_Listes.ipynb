# LISTES et ENSEMBLES

Une liste est une séquence ordonnée de valeurs. Il est indiqué par une séquence de valeurs, séparées par des virgules, entre crochets. 

Une liste peut contenir des valeurs égales. 

Une liste ne doit pas nécessairement être homogène : elle peut contenir tous les types de valeurs.

```python
# une liste de chiffres:
ln=[1,2,3,0.7,9]
# une liste de textes:
lt=['SUN','Flower','Mountain']
#mix:
lmix=[1,2,3,0.7,9,'SUN','Flower','Mountain']
```

Pour obtenir la longueur d'une liste, on utilise la commande len()

```python
len(ln)
```

```python
print(len(ln))
print(len(lt))
print(len(lmix))
```

### LISTES DES LISTES (listes imbriquées)

Une "liste de listes" est une liste dont les éléments sont eux-mêmes des listes.
On peut l'utiliser pour représenter des tableaux, des matrices, ou des données organisées par lignes et colonnes.


```python
# Exemple : une matrice 3x3
matrice = [
    [1, 2, 3],   # première ligne
    [4, 5, 6],   # deuxième ligne
    [7, 8, 9]    # troisième ligne
]

# Afficher toute la matrice
print("Matrice complète :")
print(matrice)
```

#### Accès aux éléments
Pour accéder à un élément, on utilise deux indices :
- le premier pour choisir la ligne
- le deuxième pour choisir la colonne

```python
# première ligne
matrice[0]  
```

```python
# première ligne, deuxième element
matrice[0][1]  
```

##### FLATTEN MATRIX

« Flatten » signifie transformer une matrice (une liste de listes) en une seule liste qui contient tous les éléments dans un seul vecteur (une dimension). C'est une operation tres frequente pour le nettoyage des données.

```python
matrice = [
    [1, 2, 3],
    [4, 5, 6]
]
```

Après flatten je veux obtenir listeFlat=[1,2,3,4,5,6]

Pour effectuer cette opération, il existe plusieurs procédures possibles. Je vous donne ici la formulation la plus rapide, que nous comprendrons mieux dans les prochaines séances.

```python
# cette fonctionne prend en entrée une matrice (liste de listes) et retourne une liste aplatie. 
# #On la declare au debut du code pour pouvoir l'utiliser chaque fois qu'on en a besoin.   

def flatten_matrix(matrix):
    flattened_list = [valeur for ligne in matrix for valeur in ligne]
    return flattened_list

fm=flatten_matrix(matrice)
```

```python
fm
```

## LA BIBLIOTHÈQUE NUMPY

Pour manipuler des **listes de nombres** et effectuer des **calculs statistiques** rapidement, on peut utiliser la bibliothèque **NumPy**.  

Avec NumPy, on peut facilement calculer :  
- la **moyenne** (`mean`)  
- l’**écart-type** (`std`)  
- la **médiane** (`median`)  
- les **quartiles** (`percentile`)  

```python
import numpy as np
```

```python
ll=[1,4,6,7.5,9.4]

print(np.mean(ll)) # moyenne
print(np.std(ll))  # écart-type
print(np.median(ll)) # médiane
print(np.percentile(ll, 25)) # premier quartile
print(np.percentile(ll, 75)) # troisième quartile
#!/usr/bin/env python3
```
