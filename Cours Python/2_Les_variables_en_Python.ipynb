# Type conversion

En Python, il est souvent nécessaire de **convertir** une valeur d'un type à un autre.  
Cela s'appelle le **casting** ou la **conversion de type**. Les conversions les plus courantes sont :

- **`int()`** : convertit une valeur en entier  
```python
  x = int(3.7)  # x devient 3
  y = int("10") # y devient 10



- **`float()`** : convertit une valeur en nombre à virgule flottante
```python
    x = float(5)    # x devient 5.0 
    y = float("3.14") # y devient 3.14



- **`str()`** : convertit une valeur en chaîne de caractères  
```python
  x = str(100)    # x devient "100"  
  y = str(3.14)   # y devient "3.14"



  Attention : toutes les conversions ne sont pas possibles. Par exemple, int("bonjour") provoquera une erreur.

````python
a=4.9
````

````python
b=int(a)
````

````python
b
````

````python
float(b)
````

````python
str(a)
````

````python
str(a)+str(b)
````

````python
'test'+str(b)
````

#### QUELQUE OPERATION POUR LES VARIABLES TEXTUELLES

````python
tt='La population de la France est de 67,8 millions. La population de l\'Italie est de 58,98 millions.' 
````

````python
# calculer la longueur du texte
len(tt)
````

````python
# mettre tout le teste en minuscule/majuscule
print(tt.lower())
print(tt.upper())
````

````python
#modifier certaines parties du texte
tt2=tt.replace('Italie','Argentine')
````

````python
tt
````

````python
tt2
````

````python
#On peut encheiner les operations.
tt3=tt.replace('Italie','Argentine').replace('58,98','46,01')
````

````python
tt3
````

````python
#Tester si un texte contient un mot
'Italie' in tt
````

````python
'Italie' in tt3
````

````python
#premiere lettre du texte
tt[0]
#derniere lettre du texte
tt[-1]
#sous-texte (de la position 4 a la position 14)
tt[4:15]
````

### COMPOSITION DES VARIABLES TEXTUELLES

````python
# Données issues d’une enquête par questionnaire
prenom = "Karim"
age = 21
niveau_etudes = "licence"
ville = "Lyon"
````

On veut combiner plusieurs variables textuelles afin de construire une phrase décrivant le profil d’un individu à partir de données issues d’un questionnaire sociologique (âge, sexe, profession, lieu de résidence).

````python
# Conversion de l’âge en texte pour pouvoir l’intégrer dans la phrase
age_texte = str(age)
````

````python
age_texte
````

````python
# Construction d’une phrase descriptive
phrase = prenom + " a " + age_texte + " ans, est en " + niveau_etudes + " et habite à " + ville
````

````python
print(phrase)
````

### ECRITURE SUR FICHIER 

Dans le cadre du traitement de données, il est souvent nécessaire de sauvegarder les résultats d’un programme dans un fichier (par exemple pour conserver un résumé de profils, des réponses nettoyées ou des descriptions d’individus). On suive ces étapes:

- Creation d'une variable avec le nom du fichier de sortie
- Ouverture  d’un fichier de sortie en mode écriture
- Écriture
- Fermeture du fichier de sortie


````python
#Creation d'une variable avec le nom du fichier de sortie
outfile='exemple_ecriture.txt'
````

````python
#Ouverture  d’un fichier de sortie en mode écriture
ww=open(outfile,'w')
````

Si je veux ajouter du text à un fichier pre-existent je utilise 'a' à la place de 'w':

ww=open(outfile,'a')

````python
# Ecriture
ww.write(phrase)
````

````python
#Fermeture du fichier
ww.close()
````
